```python
import pandas as pd
df  = pd.read_csv('attack data.csv')
df.head()
df["method"] = df["method"].astype(str)
df["path"] = df["path"].astype(str)
df["User-Agent"] = df["User-Agent"].astype(str)
df["Accept-Encoding"] = df["Accept-Encoding"].astype(str)
df["fulltext"] = df[["method","path","User-Agent","Accept-Encoding"]].agg(" ".join, axis=1)
df.head(100)
```

```python
import nltk
nltk.download('punkt')
```

```python
# df1.info()
df['label']=1
df
```

```python
df1 = df[['fulltext', 'label']]
df1
```

```python
import re
```

```python
sent = '""""'.join(df1['fulltext'])
cleaned_entries = [re.sub(r'[^a-zA-Z0-9.""""]', ' ', sent)]
type(cleaned_entries)
cleaned_entries
str1= str(cleaned_entries[0])
sentences = [sentence.strip() for sentence in re.split(r'"{4}', str1) if sentence.strip()]
sentences
dat=[ ]
for i in range(len(sentences)):
  rat= nltk.sent_tokenize(sentences[i])
  dat.append(rat)
dat
tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
df1['sent']=sentences
df2 = df1[['label','sent']]
df2.to_csv('output.csv', index=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
flattened_sentences = [' '.join(tokens) for tokens in tokenized_sentences]
vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(flattened_sentences)
feature_names = vectorizer.get_feature_names_out()
bow_matrix
numpy_matrix = bow_matrix.toarray()
type(numpy_matrix)
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(flattened_sentences)
numpy_matrix1 = tfidf_matrix.toarray()
numpy_matrix1
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.model_selection import train_test_split

texts_train, texts_test, labels_train, labels_test = train_test_split(numpy_matrix, df1['label'], test_size=0.2, random_state=42)

# Create a Naive Bayes classifier
nb_classifier = MultinomialNB()

# Train the classifier
nb_classifier.fit(texts_train,labels_train )

# Make predictions on the test set
predictions = nb_classifier.predict(texts_test)

# Evaluate the model
accuracy = accuracy_score(labels_test, predictions)
print(f"Accuracy: {accuracy:.2f}")
```
